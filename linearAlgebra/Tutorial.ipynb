```python
import numpy as np
import pandas as pd 
```

```python
v_1 = np.array([1,2,3])
v_2 = np.array([4,6,2])
```

```python
np.dot(v_1,v_2)
```

```python
list([v_2,v_1])
```

```python
np.matrix(list([v_2,v_1]))
```

```python
transpose=np.transpose(np.matrix(list([v_2,v_1])))
transpose
```

```python
np.matrix(list([v_2,v_1])).T
```

```python
pd.DataFrame(transpose,columns=["V_2","V_1"]).plot(figsize=(16,3))
```

```python
pd.DataFrame(transpose.T).plot(figsize=(16,3))
```

```python
pd.DataFrame(transpose*transpose.T).plot(figsize=(16,3))
```
